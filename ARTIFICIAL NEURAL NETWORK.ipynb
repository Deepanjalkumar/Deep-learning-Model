```python
#artificial neural network
np.random.seed(56)
data.head(1)
x=data.iloc[:,0:10]
y=data.iloc[:,10:11]
```

```python
#splitting the datasets into training and testing in the ratio of 0.7 and 0.3
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
#features scaliong
from sklearn.preprocessing import MinMaxScaler
mca=MinMaxScaler()
x_train=mca.fit_transform(x_train)
x_test=mca.fit_transform(x_test)
```

```python
#building artificial neural network
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LeakyReLU,PReLU,ELU
classifier=Sequential()
```

```python
#building first hidden layers
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu',input_dim=10))
```

```python
#building second hidden layesr
classifier.add(Dense(units=6,kernel_initializer='he_uniform',activation='relu'))
```

```python
#building output layer
classifier.add(Dense(output_dim=1,init='glorot_uniform',activation='sigmoid'))
```

```python
#compiling the layesr
classifier.compile(optimizer='Adamax',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fittingv the ann model
fit=classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=120)
```

```python
#predict
predict=classifier.predict(x_test)
predict=predict>0.5
```

```python
from sklearn import metrics
print(metrics.accuracy_score(predict,y_test))
```
